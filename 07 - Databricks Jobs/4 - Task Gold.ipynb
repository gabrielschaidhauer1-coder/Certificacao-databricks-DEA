```python
# Gold Layer - Agregações e Métricas de Negócio
from pyspark.sql.functions import (
    col, sum, count, avg, round, 
    date_trunc
)

# Leitura da Silver
df_silver = spark.read.table("techdados.silver.vendas")

# ========================================
# GOLD 1: Vendas por Loja
# ========================================
df_vendas_loja = (
    df_silver
    .groupBy("loja")
    .agg(
        count("id").alias("total_pedidos"),
        sum("quantidade").alias("total_itens"),
        round(sum("valor_total"), 2).alias("receita_total"),
        round(avg("valor_total"), 2).alias("ticket_medio")
    )
)

df_vendas_loja.write.format("delta").mode("overwrite").saveAsTable("techdados.gold.vendas_por_loja")
```
