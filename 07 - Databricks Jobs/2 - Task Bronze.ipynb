```python
from pyspark.sql.functions import current_timestamp

# Leitura do CSV raw
df_bronze = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .option("delimiter", ",")
    .load("/Volumes/techdados/bronze/vol_landing/vendas/")
)

# Adiciona metadados de ingestão
df_bronze = (
    df_bronze
    .withColumn("_ingestion_timestamp", current_timestamp())
    .withColumn("_source_file", df_bronze["_metadata.file_path"])
)

# Salva como Delta na camada Bronze
(
    df_bronze.write
    .format("delta")
    .mode("append")
    .option("mergeSchema", "true")
    .saveAsTable("techdados.bronze.vendas")
)
```
