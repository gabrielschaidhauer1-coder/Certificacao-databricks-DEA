## Relacionamentos

| Relação | Tipo | Descrição |
|---------|------|-----------|
| CUSTOMERS → ORDERS | 1:N | Um cliente pode ter vários pedidos |
| ORDERS → ORDER_ITEMS | 1:N | Um pedido contém vários itens |
| PRODUCTS → ORDER_ITEMS | 1:N | Um produto pode estar em vários itens |
