# Arquitetura Medallion no Databricks

## Visão Geral

A arquitetura Medallion organiza dados em um **Data Lakehouse** em três camadas progressivas de qualidade.

![Arquitetura Medallion](https://www.databricks.com/sites/default/files/inline-images/building-data-pipelines-with-delta-lake-120823.png)

---

## Camadas

| Camada | Descrição | Características |
|--------|-----------|-----------------|
| **Bronze** | Dados brutos ingeridos | Formato original, histórico completo, sem transformações |
| **Silver** | Dados filtrados e limpos | Validação, deduplicação, enriquecimento |
| **Gold** | Agregações de negócio | Métricas, KPIs, pronto para consumo |

---

## Pontos-Chave para Certificação

### Bronze (Raw)
- Ingestão de múltiplas fontes: CSV, JSON, Parquet, JDBC
- Preserva dados originais para auditoria
- Usa Auto Loader ou COPY INTO para streaming

### Silver (Cleaned)
- Aplicação de schema enforcement
- Tratamento de dados nulos e duplicados
- Joins e transformações básicas

### Gold (Business)
- Agregações otimizadas para consultas
- Tabelas dimensionais e fatos
- Consumo direto por ferramentas de BI

---

## Benefícios

- **Qualidade incremental**: dados melhoram a cada camada
- **Rastreabilidade**: histórico preservado na Bronze
- **Flexibilidade**: Silver alimenta tanto Gold quanto AI/ML
- **Performance**: Gold otimizada para consultas analíticas
