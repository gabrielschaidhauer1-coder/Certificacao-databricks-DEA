```sql
-- Criar Catalog com Managed Location
CREATE CATALOG IF NOT EXISTS techdados
MANAGED LOCATION 'abfss://techdados@deadatabricks.dfs.core.windows.net/tables'
COMMENT 'Catalog TechDados - Projeto ETL Certificação';

-- Criar Schemas (herdam o managed location do catalog)
CREATE SCHEMA IF NOT EXISTS techdados.bronze COMMENT 'Camada Bronze - Dados Raw';
CREATE SCHEMA IF NOT EXISTS techdados.silver COMMENT 'Camada Silver - Dados Limpos';
CREATE SCHEMA IF NOT EXISTS techdados.gold COMMENT 'Camada Gold - Agregações';

-- Criar Volume Landing
CREATE EXTERNAL VOLUME IF NOT EXISTS techdados.bronze.vol_landing
LOCATION 'abfss://techdados@deadatabricks.dfs.core.windows.net/raw'
COMMENT 'Volume Landing - Arquivos Raw de Ingestão';
```
