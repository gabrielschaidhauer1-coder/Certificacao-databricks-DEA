
## 🐞 Debugger Interativo em Python no Databricks

### ✅ Requisitos
Para utilizar o **debugger interativo** em notebooks Python, o notebook deve estar conectado a **um dos seguintes tipos de compute**:

- **Serverless Compute**
- **Compute com Access Mode = Standard**  
  (antigo *Shared*) — Databricks Runtime **14.3 LTS ou superior**
- **Compute com Access Mode = Dedicated**  
  (antigo *Single User*) — Databricks Runtime **13.3 LTS ou superior**
- **Compute com Access Mode = No Isolation Shared** — Databricks Runtime **13.3 LTS ou superior**

---

### ⚙️ Como habilitar o debugger

Siga os passos abaixo para ativar o debugger interativo no Databricks:

1. Clique no seu **usuário** no canto superior direito do workspace e selecione **Settings**.
2. No menu lateral de **Settings**, clique em **Developer**.
3. Em **Editor settings**, ative a opção **Python Notebook Interactive Debugger**.

Após isso, o debugger estará disponível nos notebooks Python.


Link Documentação: https://docs.databricks.com/aws/en/notebooks/debugger
