# Cluster Databricks
![Arquitetura Databricks](https://miro.medium.com/1*7sGRYZwYj9nDv71z9KC0fg.png)

Os workloads (Notebook, Job e Pipeline) se conectam a um cluster, que é composto por um Driver e vários Workers, todos executando em máquinas virtuais (VMs).
O Driver coordena a execução (planejamento, agendamento e controle), enquanto os Workers executam as tarefas distribuídas e processam os dados em paralelo.

### Cluster Serveless
A computação sem servidor no Azure Databricks é um serviço gerenciado que aloca automaticamente recursos de computação sob demanda, acelerando início e escalonamento de cargas de trabalho sem necessidade de provisionar VMs. É usada por notebooks, jobs, SQL warehouses e outros recursos do Databricks, com requisitos como Unity Catalog habilitado e disponibilidade regional. A versão é gerenciada pelo Databricks.

- Limitações Serveless
  - ❌ Não suporta R e Scala (apenas Python e SQL).

  - ❌ Não permite bibliotecas JAR personalizadas nem dependências nativas do sistema operacional.

  - ❌ Sem acesso direto à Internet para UDFs ou código customizado (somente endpoints aprovados).

  - ❌ Configurações avançadas de Spark não são suportadas (Spark conf, tuning manual, GC, etc.).

  - ❌ Logs detalhados do Spark não ficam acessíveis ao usuário.

  - ❌ Streaming tem suporte limitado (nem todos os modos e fontes funcionam).

  - ❌ Não suporta algumas fontes de dados e conectores personalizados.

  - ❌ Sem cache manual (CACHE TABLE) — cache é gerenciado automaticamente.

  - ⚠️ Requer Unity Catalog habilitado.

  - ⚠️ Menor controle de infraestrutura, em troca de simplicidade e elasticidade.


### Computação Clássica
A computação clássica (clusters) no Azure Databricks oferece controle total sobre a infraestrutura, permitindo configurar tipos de VM, versões do runtime, bibliotecas, parâmetros do Spark e acesso à rede. Ela é indicada para workloads mais complexos, como uso de Scala/R, bibliotecas JAR, streaming avançado e integrações personalizadas. Diferente do serverless, exige gerenciamento explícito de início, parada, escalabilidade e custos do cluster.



| Critério              | All-Purpose Cluster                                  | Job Cluster                                  |
| --------------------- | ---------------------------------------------------- | -------------------------------------------- |
| Provisionamento       | Criado manualmente pelo usuário                      | Criado automaticamente pelo Databricks Jobs  |
| Ciclo de vida         | Persistente      | Finalizado automaticamente ao término do job |
| Caso de uso           | Workloads interativos (desenvolvimento e exploração) | Workloads automatizados e produção           |
| Compartilhamento      | Pode ser compartilhado entre múltiplos usuários      | Isolado para um único job                    |
| Custo                 | Maior custo devido à execução contínua               | Menor custo por execução sob demanda         |
| Performance           | Pode sofrer contenção de recursos                    | Recursos dedicados ao job                    |
| Governança            | Menor isolamento entre execuções                     | Maior isolamento e previsibilidade           |
| Recomendação de uso | Usar para notebooks interativos                      | Usar para pipelines e jobs agendados         |
