# 🪄 Magic Commands no Databricks  
Os *magic commands* permitem alternar entre linguagens e executar comandos especiais dentro do notebook.

Abaixo estão os principais comandos.

- **%python, %scala, %sql, %r**: Escolhe a linguagem de programação utilizada na célula.
- **%md**: Permite escrever texto formatado em Markdown.
- **%sh**: Executa comandos de terminal (Linux Shell) diretamente do notebook.
- **%fs**: Executa comandos no sistema de arquivos do Databricks (DBFS), como `ls`, `cp`, `rm`.
- **%run**: Importa e executa outro notebook dentro do atual.
- **%pip**: Instala bibliotecas Python apenas no ambiente do notebook.

```python
%sh
ifconfig
```

```python
%sh
hostname -I
```

```python
%fs ls /databricks-datasets/
```

```python
%pip instal pandas
```
