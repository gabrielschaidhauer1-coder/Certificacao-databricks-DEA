## Questão 4 - Expectativas no Delta Live Tables

**Resposta Correta: D**

**Justificativa:**

**Opções de Constraint:**

- `ON VIOLATION FAIL UPDATE`: **Falha a task** (resposta correta)

- **A) Incorreta**: Isso seria com `DROP ROW`
- **B) Incorreta**: Registros não são adicionados
- **C) Incorreta**: Não cria campo adicional

- **E) Incorreta**: Não existe tabela de quarentena automática

## Questão 9 - Criação de UDF (User Defined Function)

**Resposta Correta: D**

**Justificativa:**  
A sintaxe correta para criar UDF em SQL é `CREATE FUNCTION` (não "CREATE UDF") e deve incluir `RETURNS` para especificar o tipo de retorno.

```sql
CREATE FUNCTION convert_distance(distance DOUBLE, measure STRING)
RETURNS DOUBLE
RETURN
CASE
  WHEN measure = "Miles" THEN (distance * 1.60934)
  WHEN measure = "KM" THEN (distance * 0.621371)
  ELSE distance
END;
```

- **A) Incorreta**: "CREATE UDF" sem FUNCTION e sem RETURNS
- **B) Incorreta**: "CREATE UDF FUNCTION" é redundante
- **C) Incorreta**: Falta especificar RETURNS DOUBLE
- **D) ✅ Correta**: CREATE FUNCTION com RETURNS é a sintaxe correta
- **E) Incorreta**: "CREATE UDF" não é válido

**Documentação:**

## Questão 10 - Higher-Order Functions

**Resposta Correta: E**

**Justificativa:**  
Para filtrar elementos de um array baseado em uma condição, usa-se a função `FILTER` com expressão lambda.

```sql
SELECT
  department_id,
  FILTER(employees, e -> e.employee_salary > 500000) AS employees_with_high_salary
FROM company;
```

**Higher-Order Functions:**
- **FILTER**: Filtra elementos que atendem condição (resposta correta)
- **TRANSFORM**: Transforma cada elemento do array
- **EXISTS**: Retorna boolean se algum elemento atende condição
- **REDUCE**: Agrega elementos em um único valor

- **A) Incorreta**: FETCH não é uma função válida
- **B) Incorreta**: TRANSFORM modifica elementos, não filtra
- **C) Incorreta**: EXIST retorna boolean, não array filtrado
- **D) Incorreta**: REDUCE agrega valores, não filtra
- **E) ✅ Correta**: FILTER é a função apropriada para filtrar arrays

**Documentação:**
- [Higher-order functions](https://learn.microsoft.com/pt-br/azure/databricks/sql/language-manual/sql-ref-functions-builtin#higher-order-functions)
- [FILTER function](https://learn.microsoft.com/pt-br/azure/databricks/sql/language-manual/functions/filter)
