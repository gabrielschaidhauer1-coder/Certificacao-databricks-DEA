## Introdução ao Databricks SQL (DBSQL)

O **Databricks SQL (DBSQL)** é um serviço otimizado para a execução de **consultas SQL** e **BI** sobre o **Data Lake**, consolidando a arquitetura **Lakehouse**.  
Para a certificação, é fundamental entender onde o DBSQL se encaixa no ecossistema Databricks.

---

### 🔹 Conceitos Chave

- **Data Intelligence Platform**
  - O DBSQL permite que **analistas de dados** e **engenheiros** trabalhem em um ambiente familiar de **SQL**
  - Elimina a necessidade de mover dados para um **Data Warehouse proprietário**

- **Motor Photon**
  - Motor de execução **vetorial em C++**
  - Acelera drasticamente as consultas SQL
  - É o **padrão** em SQL Warehouses

- **Unity Catalog**
  - Totalmente integrado ao DBSQL para **governação de dados**
  - Permite **controlo de acesso granular** usando `GRANT` e `REVOKE`

---

### 🔹 Diferença entre Workspaces

- **Persona Data Engineering**
  - Focada em:
    - Notebooks
    - Clusters tradicionais
    - Pipelines (Delta Live Tables)

- **Persona SQL**
  - Interface dedicada para:
    - Editor SQL
    - Dashboards
    - Configuração de SQL Warehouses
