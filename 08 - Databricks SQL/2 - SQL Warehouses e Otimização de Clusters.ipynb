# SQL Warehouses e Otimização de Clusters

## O que são SQL Warehouses?

SQL Warehouses são recursos computacionais dedicados exclusivamente ao Databricks SQL. Diferem dos clusters "All-Purpose" por serem otimizados para baixa latência de consulta e alta concorrência de utilizadores.

**Ponto-chave para certificação:** SQL Warehouses executam **apenas SQL** — não suportam notebooks Python, Scala ou R.

---

## Tipos de SQL Warehouse

| Tipo | Startup Time | Gestão | Recursos Principais |
|------|--------------|--------|---------------------|
| **Serverless** | 2-6 segundos | Databricks (fora da conta cloud do cliente) | Intelligent Workload Management, escalonamento instantâneo |
| **Pro** | ~2 minutos | Cliente (na conta Azure) | Photon, Predictive I/O |
| **Classic** | ~2 minutos | Cliente (na conta Azure) | Funcionalidades básicas |

### Serverless

- Compute gerido inteiramente pela Databricks
- **Não** consome recursos da subscrição Azure do cliente
- Ideal para workloads imprevisíveis ou picos de utilização
- Suporta **Intelligent Workload Management** (IWM): escalonamento baseado em carga real

### Pro

- Inclui **Photon** (motor de execução vectorizado em C++)
- **Predictive I/O**: pré-carrega dados com base em padrões de acesso
- Compute reside na conta cloud do cliente (maior controlo de rede/segurança)

### Classic

- Sem optimizações avançadas de I/O
- Opção de custo mais baixo para workloads simples

## Configuração e Sizing

### T-Shirt Sizing (Escala Vertical)

| Tamanho | Cluster Workers | Caso de Uso |
|---------|-----------------|-------------|
| 2X-Small | 1 | Desenvolvimento, testes |
| X-Small | 1 | Consultas leves |
| Small | 2 | Workloads pequenos |
| Medium | 4 | Workloads médios |
| Large | 8 | Dashboards complexos |
| X-Large | 16 | BI empresarial |
| 2X-Large | 32 | Cargas intensivas |
| 3X-Large | 64 | Data warehousing pesado |
| 4X-Large | 128 | Máxima capacidade |

**Para certificação:** Aumentar o tamanho do warehouse melhora a performance de **consultas individuais** (escala vertical).

### Multi-Cluster Load Balancing (Escala Horizontal)

Configuração de **Min/Max clusters** para lidar com concorrência:

```
Min Clusters: 1
Max Clusters: 40
```

**Comportamento:**
- Com mais utilizadores simultâneos, novos clusters são adicionados automaticamente
- Consultas são distribuídas entre clusters disponíveis
- Melhora a **concorrência**, não a velocidade de consultas individuais

**Para certificação:** 
- Escala horizontal → mais utilizadores simultâneos
- Escala vertical → consultas individuais mais rápidas

## Auto-Stop e Gestão de Custos

| Tipo | Auto-Stop Padrão | Configurável |
|------|------------------|--------------|
| Serverless | 5 minutos | Sim (5-60 min) |
| Pro/Classic | 10 minutos | Sim (10-120 min) |

**Ponto de certificação:** Warehouses parados **não consomem DBUs**. O auto-stop é essencial para controlo de custos.

---

## Photon Engine

Motor de execução nativo em C++ que acelera workloads SQL e Spark DataFrame.

### Características

- Execução **vectorizada**: processa múltiplas linhas em simultâneo
- Optimizado para operações em Delta Lake
- **Compatível** com código Spark existente (sem alterações necessárias)
- Habilitado por padrão em warehouses Pro e Serverless

### Operações Aceleradas pelo Photon

| Optimizado | Não Optimizado |
|------------|----------------|
| Aggregations (SUM, COUNT, AVG) | UDFs Python/Scala |
| Joins (todas as variantes) | Operações RDD |
| Filtros e projecções | Streaming estruturado |
| Operações Delta (MERGE, UPDATE, DELETE) | Machine Learning |
| Leitura de Parquet/Delta | Acessos a fontes externas não-Delta |

## Comparação: SQL Warehouse vs All-Purpose Cluster

| Aspecto | SQL Warehouse | All-Purpose Cluster |
|---------|---------------|---------------------|
| Linguagens | Apenas SQL | Python, Scala, SQL, R |
| Caso de uso | BI, dashboards, consultas ad-hoc | ETL, notebooks, ML |
| Optimização | Latência e concorrência | Flexibilidade |
| Photon | Pro/Serverless (incluído) | Opcional (configurável) |
| Custo (DBU) | Geralmente menor para SQL | Variável |

## Pontos Críticos para Certificação

### O que PODE fazer com SQL Warehouses

- ✅ Conectar ferramentas BI (Power BI, Tableau, Looker)
- ✅ Executar queries SQL e dashboards
- ✅ Usar SQL Editor no workspace
- ✅ Agendar queries e alertas
- ✅ Criar e consultar tabelas Delta

### O que NÃO PODE fazer

- ❌ Executar notebooks Python/Scala/R
- ❌ Treinar modelos ML
- ❌ Executar jobs Spark genéricos
- ❌ Usar Spark Streaming

### Perguntas Frequentes no Exame

1. **Cenário:** "Vários analistas reportam lentidão ao executar dashboards simultaneamente."
   **Resposta:** Aumentar o número máximo de clusters (escala horizontal)

2. **Cenário:** "Uma consulta individual está lenta."
   **Resposta:** Aumentar o tamanho do warehouse (escala vertical)

3. **Cenário:** "Precisa conectar Power BI ao Databricks."
   **Resposta:** Usar SQL Warehouse (não All-Purpose Cluster)

4. **Cenário:** "Quer reduzir custos quando não há utilizadores activos."
   **Resposta:** Configurar Auto-Stop com tempo adequado

5. **Cenário:** "Precisa de startup instantâneo para workloads imprevisíveis."
   **Resposta:** Usar Serverless SQL Warehouse
