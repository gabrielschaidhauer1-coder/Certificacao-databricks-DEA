# Versionamento no Databricks - Guia Completo com Fontes Oficiais Azure

## 📋 Histórico de Versão do Notebook (Notebook Version History)

### O que é?
O Azure Databricks mantém automaticamente um histórico de versões dos notebooks, permitindo visualizar e restaurar snapshots anteriores. 

### Funcionalidades

- **Salvamento Automático**: Cada mudança no notebook é automaticamente salva no histórico de revisões
- **Visualizar Versões**: Acesso ao histórico através do painel lateral direito
- **Restaurar Versões**: Possibilidade de restaurar qualquer versão anterior como a versão mais recente 
- **Adicionar Descrições**: Salvar versões com descrições personalizadas
- **Deletar Versões**: Remover versões específicas do histórico
- **Limpar Histórico**: Opção de limpar todo o histórico (o histórico de versões não pode ser recuperado após ser limpo) 

### Limitações Principais

1. **Escopo Limitado**: O versionamento vive apenas no ambiente Databricks. Se o ambiente for deletado, todo o trabalho e histórico de versões são perdidos
2. **Sem Ramificação (Branching)**: Não suporta branches para desenvolvimento paralelo
3. **Sem CI/CD**: Não integra com pipelines de integração/entrega contínua
4. **Notebook Individual**: Controla apenas um notebook por vez, não gerencia projetos completos

## 🔄 Pastas Git do Databricks / Repositórios (Databricks Git Folders / Repos)

### O que é?
Integração de repositórios Git dentro do workspace Databricks para controle de versão, colaboração e workflows de CI/CD. 

### Funcionalidades Principais

#### Controle de Versão Completo
- Suporte a operações Git padrão: clone, commit, push, pull, merge, rebase
- Gerenciamento de branches: criar, alternar, mesclar branches
- Resolução de conflitos através da interface do Databricks
- Histórico Git permanente e completo

#### Colaboração em Equipe
- Usuários podem criar pull requests após fazer push das mudanças
- Cada membro da equipe trabalha em sua própria pasta Git mapeada para o repositório remoto
- Code review através do provedor Git
- Trabalho paralelo usando feature branches

#### Integração CI/CD
- Repos API permite automatizar atualizações de pastas Git quando pull requests são mesclados
- Integração com ferramentas CI/CD como GitHub Actions, Azure DevOps, Jenkins
- Pastas Git de produção servem propósito diferente das pastas de usuário

### Provedores Git Suportados
GitHub, GitLab, Azure DevOps, Bitbucket, AWS CodeCommit 

### Localização no Workspace
- **User-level folders**: `/Workspace/Users/<email>/` - onde usuários desenvolvem e fazem push de mudanças
- **Production folders**: Criadas por admins fora das pastas de usuário, contêm branches de deployment de produção

## 🔧 Configuração de Repositórios Git no Azure

### Pré-requisitos
Git folders habilitados no workspace (habilitado por padrão), conta em provedor Git (GitHub, GitLab, Azure DevOps, Bitbucket ou AWS CodeCommit), personal access token (PAT) ou credenciais OAuth para repositórios privados. 

### Passos de Configuração
1. **Configurar Credenciais Git**
   - Clicar no nome de usuário no topo do workspace Azure Databricks → Settings → Linked accounts → Add Git credential 
   - Selecionar provedor Git
   - Alguns provedores oferecem vinculação de conta OAuth, enquanto outros requerem personal access token (PAT) 

2. **Clonar Repositório**
   - Navegar para a seção Repos
   - Add Repo → Clone from Git URL
   - Selecionar branch desejado

3. **Operações Git**
   - Criar/alternar branches
   - Fazer commits
   - Push/Pull mudanças
   - Resolver conflitos

### Segurança no Azure
Databricks Git folders incluem recursos de segurança: usar Azure Key Vault para criptografar tokens Git com suas próprias chaves de criptografia (customer managed keys), e admins do workspace podem restringir quais repositórios remotos os usuários podem acessar. 

### Controle de Acesso
Se você usa autenticação Microsoft Entra ID com Azure DevOps, a lista de permissões padrão restringe URLs Git

## 🚀 CI/CD com Azure Databricks Repos

### Fluxo Recomendado

Usuários podem criar pull requests após fazer push. Quando o PR é mesclado, automação puxa as mudanças para as pastas Git de produção usando a API Repos do Azure Databricks. 

### Configuração de Pastas de Produção

Pastas Git de produção servem propósito diferente das pastas de nível de usuário localizadas em /Workspace/Users/. Pastas de usuário atuam como checkouts locais onde usuários desenvolvem e fazem push de mudanças. Pastas de produção são criadas por admins Databricks fora das pastas de usuário e contêm branches de deployment de produção. 

### Permissões para Produção

Para pastas Git de produção, limitar acesso de usuário para apenas execução (run-only) e permitir apenas admins e service principals do Azure Databricks editarem. 

### Ferramentas CI/CD com Azure

Usar ferramentas CI/CD externas como GitHub Actions para puxar os commits mais recentes para uma pasta Git de produção quando um pull request é mesclado no branch de deployment.

## ✅ Melhores Práticas (Best Practices)

### Para Histórico de Notebook
1. Usar apenas para trabalho exploratório
2. Não confiar como única forma de backup
3. Migrar para Git quando o projeto amadurecer

### Para Azure Databricks Repos
1. Cada desenvolvedor deve trabalhar em sua própria branch
2. Limitar acesso a pastas Git de produção para apenas run-only, permitindo apenas admins e service principals do Azure Databricks editarem 
3. Apenas um usuário deve realizar operações Git em cada pasta Git
4. Usar feature branches para desenvolvimento
5. Fazer pull requests para revisão de código
6. Automatizar deploys para produção via CI/CD com Azure DevOps ou GitHub Actions
7. Usar service principals para automação

## 🎓 Pontos-Chave para Certificação Databricks Data Engineer Associate

1. **Repos é ESSENCIAL** para projetos de produção e colaboração em equipe
2. **Revision History** é backup temporário, não substitui Git
3. **CI/CD requer Repos** - não funciona com Workspace normal
4. **Branching strategy** só disponível com Git integration
5. Localização: **Repos** = `/Repos/`, **Workspace** = `/Workspace/`
6. Production folders devem ser atualizadas apenas programaticamente via Repos API
7. Azure Key Vault pode ser usado para segurança de credenciais Git
8. Integração nativa com Azure DevOps para CI/CD
