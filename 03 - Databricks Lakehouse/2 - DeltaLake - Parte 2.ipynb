# Managed vs External - Unity Catalog

## 🗂️ Managed Tables/Volumes
- **Databricks gerencia dados + metadados**. `DROP TABLE` = deleta tudo permanentemente.
- Ideal para dados internos e temporários (silver/gold layers).

## 🔗 External Tables/Volumes
- **Você gerencia o storage (ADLS Gen2)**. `DROP TABLE` = remove apenas metadados, dados permanecem no Azure Storage.
- Ideal para dados compartilhados e permanentes (bronze layer, raw data).
---
Managed = Databricks controla tudo | External = Você controla ADLS
---
- O Databricks recomenda tabelas gerenciadas e volumes para a maioria dos casos de uso, pois eles permitem que você aproveite ao máximo os recursos de governança do Catálogo do Unity e as otimizações de desempenho.

- 📚 [Docs: Azure Databricks - Tabelas Gerenciadas](https://learn.microsoft.com/pt-br/azure/databricks/data-governance/unity-catalog/)

```python
%sql
CREATE CATALOG IF NOT EXISTS dea_databricks
MANAGED LOCATION 'abfss://bronze@deadatabricks.dfs.core.windows.net/catalog';
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS dea_databricks.bronze
```

```python
%sql
CREATE TABLE dea_databricks.bronze.produtos (
    produto_id INT,
    nome VARCHAR(100),
    categoria VARCHAR(50),
    preco DECIMAL(10,2),
    estoque INT
);
```

```python
%sql
select * from dea_databricks.bronze.produtos
```
