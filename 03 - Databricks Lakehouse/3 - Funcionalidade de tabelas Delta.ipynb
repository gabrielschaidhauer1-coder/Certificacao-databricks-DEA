# Funcionalidade de tabelas Delta

## 🕐 Time Travel
**Auditar e consultar versões anteriores dos dados:**
```sql
-- Ver histórico de alterações
DESCRIBE HISTORY my_table;

-- Consultar versão específica (por timestamp)
SELECT * FROM my_table TIMESTAMP AS OF "2024-01-01";

-- Consultar versão específica (por número)
SELECT * FROM my_table VERSION AS OF 36;
SELECT * FROM my_table@v36;

-- Restaurar versão anterior
RESTORE TABLE my_table TO VERSION AS OF 36;
```
📚 [Docs: Time Travel](https://learn.microsoft.com/pt-br/azure/databricks/delta/history)

---

## ⚡ OPTIMIZE - Compactação
**Combinar arquivos pequenos em arquivos maiores** para melhor performance:
```sql
-- Compactar tabela
OPTIMIZE my_table;
```
**Benefício:** Reduz o número de arquivos pequenos, melhora velocidade de leitura.

📚 [Docs: OPTIMIZE](https://learn.microsoft.com/pt-br/azure/databricks/delta/optimize)

---

## 📊 Z-ORDER - Indexação
**Co-localizar informações de colunas** para queries mais rápidas:
```sql
-- Otimizar com Z-ORDER em coluna específica
OPTIMIZE my_table ZORDER BY (column_name);

-- Exemplo: otimizar por ID
OPTIMIZE my_table ZORDER BY (ID);
```
**Benefício:** Queries com filtros nessas colunas ficam muito mais rápidas.

📚 [Docs: Z-ORDER](https://learn.microsoft.com/pt-br/azure/databricks/delta/data-skipping)

---

## 🧹 VACUUM - Limpeza
**Remover arquivos não utilizados** (arquivos antigos após OPTIMIZE, versões antigas):
```sql
-- Limpar arquivos não usados (padrão: 7 dias)
VACUUM my_table;

-- Limpar com período de retenção customizado
VACUUM my_table RETAIN 168 HOURS; -- 7 dias
```
⚠️ **Atenção:** Após VACUUM, pode-se perder a capacidade de restaurar versões antigas de tabelas

📚 [Docs: VACUUM](https://learn.microsoft.com/pt-br/azure/databricks/delta/vacuum)
📚 [Docs: Recomendações VACUUM](https://learn.microsoft.com/pt-br/azure/databricks/sql/language-manual/delta-vacuum?source=recommendations)

## 🎯 Resumo Rápido

| Comando | Objetivo | Quando Usar |
|---------|----------|-------------|
| `DESCRIBE HISTORY` | Ver histórico de alterações | Auditoria, troubleshooting |
| `VERSION AS OF` | Consultar versão antiga | Comparar dados, recuperar info |
| `RESTORE TABLE` | Reverter para versão anterior | Rollback de erros |
| `OPTIMIZE` | Compactar arquivos pequenos | Após muitos inserts/updates |
| `ZORDER BY` | Indexar coluna | Queries com filtros frequentes |
| `VACUUM` | Limpar arquivos antigos | Reduzir custos de storage |
