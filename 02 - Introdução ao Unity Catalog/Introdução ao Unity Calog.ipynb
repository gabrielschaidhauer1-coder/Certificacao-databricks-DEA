# Unity Catalog

## Definição

**Unity Catalog** é um catálogo de dados centralizado que fornece:
- ✅ Controle de acesso
- ✅ Auditoria automática
- ✅ Linhagem de dados
- ✅ Monitoramento de qualidade
- ✅ Descoberta de dados

**Através de múltiplos workspaces do Azure Databricks**

---

## Princípios Fundamentais

### 1. **Define Once, Secure Everywhere**
> "Defina uma vez, proteja em todo lugar"

Uma política de acesso → Aplicada em todos os workspaces da região

### 2. **Modelo de Segurança SQL ANSI**
```sql
GRANT SELECT ON TABLE catalog.schema.table TO user;
```
Sintaxe familiar para DBAs e engenheiros de dados

### 3. **Auditoria e Linhagem Integradas**
- Logs automáticos de quem acessou o quê
- Rastreamento de como dados foram criados
- Histórico completo de transformações

### 4. **Descoberta de Dados**
- Tags e documentação nos ativos
- Interface de busca
- Catálogo centralizado e pesquisável

## Namespace de 3 Níveis

### Sintaxe
```
catalog.schema.objeto
```

### Exemplos
```sql
prod.silver.customers
dev.bronze.raw_events
analytics.gold.kpi_vendas
```

### Uso em Queries
```sql
-- Referência completa
SELECT * FROM prod.silver.customers;

-- Usar catalog padrão
USE CATALOG prod;
SELECT * FROM silver.customers;

-- Usar catalog + schema padrão
USE CATALOG prod;
USE SCHEMA silver;
SELECT * FROM customers;
```

## Componentes do Unity Catalog

### **Metastore**
- **O que é:** Contêiner de nível superior para metadados
- **Escopo:** Um por região do Azure
- **Função:** Registra metadados e permissões de todos os ativos
- **Diferença do Hive:** Não é um service boundary, é multi-tenant

### **Catalog** (Nível 1)
- **O que é:** Primeiro nível de organização lógica
- **Uso típico:** Separar ambientes ou unidades de negócio
- **Exemplos:**
  - `dev` → Desenvolvimento
  - `staging` → Homologação
  - `prod` → Produção
  - `hr` → Recursos Humanos
  - `finance` → Financeiro

### **Schema / Database** (Nível 2)
- **O que é:** Segundo nível de organização
- **Uso típico:** Casos de uso, projetos, ou camadas medallion
- **Exemplos:**
  - `bronze` → Dados brutos
  - `silver` → Dados validados
  - `gold` → Dados agregados
  - `analytics` → Análises específicas

### **Objetos de Dados** (Nível 3)

#### Tables (Tabelas)
- Coleções de dados organizados em linhas e colunas
- **Managed:** Unity Catalog gerencia dados + metadados
- **External:** Unity Catalog gerencia apenas acesso

#### Views (Visualizações)
- Queries salvas sobre tabelas
- Não armazenam dados fisicamente
- **Dynamic views:** Suportam row/column-level security

#### Volumes
- Armazenam dados não-tabulares (arquivos)
- Suporta qualquer formato (JSON, XML, imagens, PDFs, etc.)

#### Functions (Funções)
- UDFs (User-Defined Functions)
- Lógica reutilizável em SQL, Python, Scala

#### Models (Modelos ML)
- Modelos de Machine Learning registrados via MLflow
- Versionamento e governança de modelos

## Objetos de Segurança

### Storage Credentials + External Locations

![External Locations Architecture](https://learn.microsoft.com/en-us/azure/databricks/_static/images/unity-catalog/external-locations-overview.png)

**Fonte:** [Microsoft Learn - Unity Catalog](https://learn.microsoft.com/en-us/azure/databricks/data-governance/unity-catalog/)

### **Storage Credentials**
- Encapsulam credenciais de cloud de longo prazo
- Fornecem acesso ao cloud storage (Azure ADLS, Blob, etc.)
- Objetos seguros do Unity Catalog

### **External Locations**
- Referenciam:
  1. Caminho no cloud storage
  2. Storage Credential necessária
- Usadas para criar external tables/volumes
- Múltiplas external locations podem usar a mesma credential

### Exemplo Visual do Diagrama
```
Storage Credential 1
    ├── External Location A → bucket/tables/raw/*
    └── External Location B → bucket/tables/processed/*

Storage Credential 2
    └── External Location C → bucket/secure/*
```
